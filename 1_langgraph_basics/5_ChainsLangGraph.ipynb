```python
## How to use chat messages as our graph State

### Messages

We can use messages whih cab be used to capture different roles within a conversation.
LandChain has various message types including 
- HumanMessage, (from the user)
- AIMessage, (from the chat model)
- SystemMessage, (for the chat model to instruct behaviour)
- ToolMessage, (from the tool call)

Every message have these important components.
- content - content of the message
- name - Specify the name of author
- response_metadata - optionally, a dict of metadata(e.g., often populated by model provider for AIMessage)
```
